### Vehicle Group – Total Number of Casualties

```python
df.columns
```

```python
vehicle_casualties = (

    df.groupby('vehicle_type')['number_of_casualties'].sum()
    .reset_index()
)
vehicle_casualties
```

```python
def vehicle(v):
    if any(x in v for x in ['Taxi/Private hire car', 'Car']):
        return 'Car'
    elif any(x in v for x in ['Motorcycle 125cc and under','Motorcycle 50cc and under', 
                              'Motorcycle over 125cc and up to 500cc', 'Motorcycle over 500c']):
        return 'Bikes'
    elif any(x in v for x in ['Bus or coach (17 or more pass seats)', 'Minibus (8 - 16 passenger seats)']):
        return 'Bus'
    elif any(x in v for x in ['Goods 7.5 tonnes mgw and over', 
                              'Goods over 3.5t. and under 7.5t', 'Van / Goods 3.5 tonnes mgw or under']):
        return 'Van'
    elif any(x in v for x in ['Agricultural vehicle']):
        return 'Agricultural vehicle'
    else:
        return 'Other vehicles'

# Create New Column:

df['vehicle_group'] = df['vehicle_type'].map(vehicle)
```

```python
vehicle_casualties = (

    df.groupby('vehicle_group')['number_of_casualties'].sum()
)
vehicle_casualties.sort_values(ascending = False).to_frame()
```

### Area – wise Percentage(%) and Total Number of Casualties

```python
area_total = (
    df.groupby('urban_or_rural_area')['number_of_casualties'].sum()
)

area_percentage = round((area_total / area_total.sum()) * 100,2).astype(str) + ' %'


df_area = pd.concat([area_total, area_percentage], axis = 1)
df_area
```

### Count of Casualties By Light Conditions

```python
df.light_conditions.value_counts()
```

```python
def light_mapping(l):
    if any(x in l for x in ['Daylight']):
        return 'Day'
    else:
        return 'Night'

df['light_cond'] = df['light_conditions'].map(light_mapping)
```

```python
light_cond = (

    df.groupby('light_cond')['number_of_casualties'].count()
    .sort_values(ascending = False)
    .astype(str) + ' - Nos of People has been affected'
)
light_cond.to_frame(name = 'Count of Casualties')
```

### Percentage (%) and Segregation of Casualties by Different Light Conditions

```python
light_cond = (
    df.groupby('light_cond')['number_of_casualties'].sum()
)

percentage = round((light_cond / light_cond.sum()) * 100,2).astype(str) + ' %'

df_light = pd.concat([light_cond, percentage], axis = 1, 
                     keys = ['Total Nos of Casualties', 'Percentage % '])
df_light
```
